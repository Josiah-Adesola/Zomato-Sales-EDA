```python
#This is used to join or merge the two tables
final_df = pd.merge(df, df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
final_df.shape
```

```python
## To check data types

final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
## PIE CHART
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.3f%%")
```

```python
final_df.columns
```

```python
#The groupby() - helps to group different columns or features together
#The size() - used to count the columns
#reset_index() - This is used to convert the data to a  dataframe format
# rename(columns=) - This is used to rename columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## Observations
1. Ratings between 4.5 - 4.9 is excellent 
2. Ratings between 4.0 to 4.4 it's considered very good
3. Most people did not rate the app at all, they were 2148 people taking most of the population

```python
ratings['Rating Count'].mean()
```

```python
ratings.head()
```

```python
ratings.describe()
```

```python
#Let's visualize our rating
plt.rcParams['figure.figsize'] = (12, 6)  #to change the size of the code below
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color',data=ratings)
```

```python
#Mapping colors to what's on the data set
plt.rcParams['figure.figsize'] = (12, 6) 
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color',data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observations
1. The non-rated has the highest, as clearly displayed in the graph
2. The maximum ratings are between 2.4 to 3.4
3. If there are missing values, one could easily use between the values stated in no2.

```python
## Count PLot
sns.countplot(x="Rating color", data=ratings,palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

## Count PLOT
The count shows the frequency of the rating color in the data frame, not the rating count actually.

Take note of that
